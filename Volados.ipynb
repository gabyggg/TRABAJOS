## VOLADOS

# Distribución binomial, para una muestra de 30, con una probabilidad de exito de 0.5 y  size es la probabilidad de éxito en cada ensayo.

```r
volados30 <- rbinom(n= 30,size = 1, prob = 0.5)
```

# un renglón y 2 columnas esta instrucción tiene aplicación en R, y muestra dos resultados en este caso es una columna con 3 renglones.

```r
layout(matrix(1:3, 1,3))
```

```r
options(repr.plot.width=12,repr.plot.height=12/1.61803)
hist(volados30, main="Histograma 30 Volados", col="red", border="blue", labels= T, ylab="Frecuencia",)
```

# Distribución binomial, para una muestra de 300, con una probabilidad de exito de 0.5 y size es la probabilidad de éxito en cada ensayo.

```r
volados300<- rbinom(n= 300,size = 1, prob = 0.5)
```

```r
options(repr.plot.width=12,repr.plot.height=12/1.61803)
hist(volados300, main="Histograma de 300 Volados", col="pink", border="blue", labels= T, ylab="Frecuencia", ylim=c(0,160), xlim=c(0,1))
```

# Distribución binomial, para una muestra de 3000,000 con una probabilidad de exito de 0.5 y size es la probabilidad de éxito en cada ensayo.

```r
volados3millones<- rbinom(n= 3000000,size = 1, prob = 0.5)
```

```r
options (repr.plot.width=12, repr.plot.height=12/1.61803)
hist(volados3millones, main="Histograma de 3 Millones de Volados", col="blue", border="red", labels=T, ylab="Frecuencia", ylim=c(0,1550e3), xlim=c(0,1))
```
