##  Generar una muestra a partir de la función Semilla con 50 datos:

```r
set.seed(555)
muestra1 <- rnorm(50,10,0.5)
summary(muestra1)
```

Muestra de 50 millones

```r
muestra2 <- rnorm(5e6,10,0.5)
summary(muestra2)
```

Función para presentar los histogramas en una matriz

```r
layout(matrix(2:1,1,2)) para rstudio
```

```r
Histograma de Muestra de 50 datos
```

```r
options(repr.plot.width=14,repr.plot.height=14/1.61803)
hist (muestra1, main="Histograma 1", col="pink", border="blue", labels= T, ylab="Muestra 1 50 ", ylim=c(0,25), xlim=c(8,12),)
```

Histograma de Muestra de 50 millones de datos

```r
options(repr.plot.width=14,repr.plot.height=14/1.61803)
hist (muestra2, main="Histograma 2", col="blue", border="pink", labels= T, ylab="Muestra 2 de 50 millones", ylim=c(0,800e3), xlim=c(7,13),)
```

```r
dif <- abs (mean (muestra1)-mean(muestra2)) / mean(muestra2)
```

```r
dif*100
```

```r
dif2 <- abs (sd (muestra1)-sd(muestra2)) / sd(muestra2)
```

```r
dif2*100
```

Conclusión podemos ver que en una distribución normal los resultados estadísticos de una población son muy cercanos a los de una muestra.
