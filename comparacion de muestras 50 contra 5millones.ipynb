##  Generar una muestra a partir de la función Semilla con 50 datos:

```r
set.seed(555)
muestra1 <- rnorm(50,10,0.5)
summary(muestra1)
```

## Muestra de 50 millones

```r
muestra2 <- rnorm(5e6,10,0.5)
summary(muestra2)
```

# Función para presentar los histogramas en una matriz

```r
layout(matrix(2:1,1,2))
```

```r
options(repr.plot.width=10,repr.plot.height=10/1.61803)
hist (muestra1, main="Histograma 1", col="pink", border="blue", labels= T, ylab="Muestra 1",)
```

```r
options(repr.plot.width=12,repr.plot.height=12/1.61803)
hist (muestra2, main="histograma 2", col="blue", border="pink", labels= T, ylab="Muestra 2",)
```

```r
dif <- abs (mean (muestra1)-mean(muestra2)) / mean(muestra2)
```

```r
dif*100
```

```r
dif2 <- abs (sd (muestra1)-sd(muestra2)) / sd(muestra2)
```

```r
dif2*100
```
